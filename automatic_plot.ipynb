```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import scipy.linalg
from scipy import stats

logistic_baseline_path = 'plots_4/MNIST/logistic_baseline_MNIST/logistic_baseline_probs_benign.npy'
logistic_baseline = np.load(logistic_baseline_path)
kernel_logistic_baseline = stats.gaussian_kde(logistic_baseline, bw_method=0.5)
    
logistic_diverse_path = 'plots_4/MNIST/logistic_diverse_MNIST/logistic_diverse_probs_benign.npy'
logistic_diverse = np.load(logistic_diverse_path)
kernel_logistic_diverse = stats.gaussian_kde(logistic_diverse, bw_method=0.5)

softmax_baseline_path = 'plots_4/MNIST/softmax_baseline_MNIST/softmax_baseline_benign.npy'
softmax_baseline = np.load(softmax_baseline_path)    
kernel_softmax_baseline = stats.gaussian_kde(softmax_baseline, bw_method=0.5)

tanh_16_diverse_path = 'plots_4/MNIST/tanh_16_diverse_MNIST/tanh_16_diverse_probs_benign.npy'
tanh_16_diverse = np.load(tanh_16_diverse_path)
kernel_tanh_16_diverse = stats.gaussian_kde(tanh_16_diverse, bw_method=0.5)   

tanh_32_diverse_path = 'plots_4/MNIST/tanh_32_diverse_MNIST/tanh_32_diverse_benign.npy'
tanh_32_diverse = np.load(tanh_32_diverse_path)
kernel_tanh_32_diverse = stats.gaussian_kde(tanh_32_diverse, bw_method=0.5)       

Tanh_baseline_16_path = 'plots_4/MNIST/Tanh_baseline_16_MNIST/Tanh_baseline_16_probs_benign.npy'
Tanh_baseline_16 = np.load(Tanh_baseline_16_path)
kernel_Tanh_baseline_16 = stats.gaussian_kde(Tanh_baseline_16, bw_method=0.5)          

```

```python
# plt.figure(figsize=(10,10))
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111)
ax.plot(np.arange(0, 1, .01), kernel_logistic_baseline.pdf(np.arange(0, 1, .01)), linewidth=4, label='Logistic')
ax.plot(np.arange(0, 1, .01), kernel_softmax_baseline.pdf(np.arange(0, 1, .01)), linewidth=4, label='Softmax')
ax.plot(np.arange(0, 1, .01), kernel_logistic_diverse.pdf(np.arange(0, 1, .01)), linewidth=4, label='LEns10')
ax.plot(np.arange(0, 1, .01), kernel_Tanh_baseline_16.pdf(np.arange(0, 1, .01)), linewidth=4, label='Tanh16')
ax.plot(np.arange(0, 1, .01), kernel_tanh_16_diverse.pdf(np.arange(0, 1, .01)), linewidth=4, label='TEns16')
ax.plot(np.arange(0, 1, .01), kernel_tanh_32_diverse.pdf(np.arange(0, 1, .01)), linewidth=4, label='TEns32')
ax.legend(fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=15)
plt.ylim(0,20)
# plt.show()
plt.savefig('mnist_benign_U_4.eps', format='eps')
```
